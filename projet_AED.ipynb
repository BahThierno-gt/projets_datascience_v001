#### NB: ***Les types de paiement sont classifies ainsi qu'il suit:***
* Selon le dictionnaire des donnees, les types de paiement sont codees de la maniere suivante:
  
1 = Credit Card;
2 = Cash; 
3 = No charge; 
4 = Dispute; 
5 = Inconnu; 
6 = Voide trip

```python
# Le nombre de fois chaque type de paiement est représenté dans les données ?
df[['payment_type']].value_counts().reset_index(name='count')
```

```python
# Calcul du pourboire moyen (la moyenne des pourboire) pour les voyages payés par carte de crédit ?

fltr_cc = df['payment_type']==1                             # Filtrer par carte de credit
prb_moy = df[fltr_cc]['tip_amount'].mean().round(2)         # Calculer du pourb. moyen pour les voy. par cartes de credits(fltr_cc)

print('Le pourboire moyen pour les voyages par carte de credit est:', prb_moy)
```

```python
# Calcul du pourboire moyen (la moyenne des pourboire) pour les voyages payés par cash ?

fltr_cash = df['payment_type']==2                            # Filtrer par carte de credit
prb_moy = df[fltr_cash]['tip_amount'].mean().round(2)         # Calculer du pourb. moyen pour les voy. par cash(fltr_cash)

print('Le pourboire moyen pour les voyages par cash est:', prb_moy)
```

```python
# Determiner le nombre de fois chaque vendeur est représenté dans les données
df['VendorID'].value_counts().reset_index(name='count')
```

```python
# Calcul de la moyenne du montant total par vendeur
df.groupby('VendorID').mean(numeric_only=True)[['total_amount']].round(2)
```

```python
# Determiner le nombre de passager ayant paye par carte de credit
# D'abord filtrer les donnees pour les voyages payes par credit card
credit_card = df[df['payment_type']==1]
# Ensuite appliquer le mask sur la colonne des passagers
credit_card['passenger_count'].value_counts().reset_index(name='count')
```

```python
# Determiner la moyenne des pourboires pour chaque passager pour les credits card uniquement
credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']].round(2)
```
